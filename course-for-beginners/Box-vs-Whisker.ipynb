```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10, 71, 10)
x = np.append(x, 120)

plt.boxplot(x, notch = False, vert = True, widths = 0.1, labels=["First box"], patch_artist=True
            , showmeans = True, sym='g+',
           boxprops=dict(color = 'red'),
           capprops = dict(color = 'red'),
           whiskerprops= dict(color = 'red'),
           flierprops=dict(color='red'))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(172371)

x = np.random.randint(10, 50, size=(50))
x1 = np.random.randint(10, 50, size=(50))
x2 = np.random.randint(10, 50, size=(50))

y = [x, x1, x2]

plt.boxplot(y, notch = False, vert = True, widths = 0.1, patch_artist=True
            , showmeans = True,
           boxprops=dict(color = 'red'),
           capprops = dict(color = 'red'),
           whiskerprops= dict(color = 'red'),
           flierprops=dict(color='red'),
           labels=["first", "secnd", "third"])
plt.show()
```
