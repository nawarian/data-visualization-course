```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(12381723)
height = np.random.randint(160, 198, size=(200))

plt.hist(height, bins=range(160, 200, 5), align='left', edgecolor='black', color='green', alpha=0.5, linewidth=1)

plt.title("My random ages")
plt.xlabel("Height")
plt.ylabel("Quantity")
plt.xticks(range(160, 200, 5))
plt.yticks(range(1, 50, 5))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import statistics

import seaborn as sns
sns.set_theme()

np.random.seed(12381723)
height = np.random.randint(160, 198, size=(200))

plt.hist(height, bins=50, color='c', edgecolor='k')

m = statistics.mean(height)
sd = statistics.stdev(height)

plt.axvline(m, color='k', linestyle='dashed')

plt.axvline(m+sd, color='gold', linestyle='dashed')
plt.axvline(m-sd, color='gold', linestyle='dashed')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import statistics
import seaborn as sns

sns.set_theme()

blood_sample_men = np.random.randint(50, 143, size=(200))
blood_sample_women = np.random.randint(50, 143, size=(200))
```

```python
plt.hist([blood_sample_men, blood_sample_women], bins=[80, 100, 125, 150], color=['green', 'orange'], label=['men', 'women'])
plt.legend()
plt.xlabel("Sugar level")
plt.ylabel("No. of samples")

plt.title("Blood sugar chart")

plt.show()
```

### Patches in Histogram

```python
n, bins, patches = plt.hist([blood_sample_men, blood_sample_women], bins=[80, 100, 125, 150], color=['green', 'orange'], label=['men', 'women']
         , rwidth=.95, orientation='vertical')

plt.legend()
plt.xlabel("Sugar level")
plt.ylabel("No. of samples")

plt.setp(patches[1][2], color='cyan')
plt.title("Blood sugar chart")

plt.show()
```
