```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11, 1)
y = np.array([2, 3, 2, 5, 1, 4, 9, 2, 6, 3])

plt.scatter(x, y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Basic scatter plot')

plt.show()
```

```python
np.random.seed(12381273)

x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

colors = np.random.rand(100)
sizes = colors * 1000

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.colorbar()

plt.show()
```
