```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-v0_8")

np.random.seed(1283128)
minutes = np.arange(1, 10, 1)
player1 = np.array([1, 2, 3, 3, 4, 4, 4, 4, 5])
player2 = np.array([1, 1, 1, 1, 2, 2, 2, 3, 4])
player3 = np.array([1, 1, 1, 2, 2, 2, 3, 3, 3])

plt.stackplot(minutes, player1, player2, player3,
              labels=["Player 1", "Player 2", "Player 3"],
              colors=["#6d904f", "#fc4f30", "#008fd5"])
plt.legend(loc="upper left")

plt.title("My awesome stack plot")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-v0_8")

np.random.seed(1283128)
days = np.arange(1, 10, 1)
dev1 = np.array([8, 6, 5, 5, 4, 2, 1, 1, 0])
dev2 = np.array([0, 1, 2, 2, 2, 4, 4, 4, 4])
dev3 = np.array([0, 1, 1, 1, 2, 2, 3, 3, 4])

plt.stackplot(days, [dev1, dev2, dev3],
              labels=["Dev 1", "Dev 2", "Dev 3"],
              colors=["#6d904f", "#fc4f30", "#008fd5"])
plt.legend(loc="upper right")

plt.title("My awesome stack plot")
plt.tight_layout()
plt.show()
```
