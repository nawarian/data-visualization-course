```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-v0_8")

np.random.seed(12371273)

y1 = np.random.randint(10, 90, size=(4))
y2 = np.random.randint(10, 90, size=(4))
y3 = np.random.randint(10, 90, size=(4))
y4 = np.random.randint(10, 90, size=(4))

x = ['Team A', 'Team B', 'Team C', 'Team D']

plt.bar(x, y1, color='r')
plt.bar(x, y2, color='b', bottom=y1)
plt.bar(x, y3, color='y', bottom=y1+y2)
plt.bar(x, y4, color='g', bottom=y1+y2+y3)

plt.xlabel("Teams")
plt.ylabel("Score")

plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 rounds")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-v0_8")

np.random.seed(12371273)

y1 = np.random.randint(10, 90, size=(4))
y2 = np.random.randint(10, 90, size=(4))
y3 = np.random.randint(10, 90, size=(4))
y4 = np.random.randint(10, 90, size=(4))

x = ['Team A', 'Team B', 'Team C', 'Team D']

plt.barh(x, y1, color='r')
plt.barh(x, y2, color='b', left=y1)
plt.barh(x, y3, color='y', left=y1+y2)
plt.barh(x, y4, color='g', left=y1+y2+y3)

plt.xlabel("Teams")
plt.ylabel("Score")

plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 rounds")
plt.show()
```
